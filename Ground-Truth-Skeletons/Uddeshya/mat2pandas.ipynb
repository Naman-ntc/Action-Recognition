```python
import numpy as np
import scipy.io as sio
import pandas as pd
```

```python
def matfile2pandas(fname):
    mat_pd = {}
    m_contents = sio.loadmat(fname)
    m1 = m_contents['joints3d']
    label = m1['label'][0][0][0][0]
    mat_pd['label'] = label
    aframes = m1['frames']
    #print(len(aframes[0][0]))
    if len(aframes[0][0])>0:
        aframes = aframes[0][0][0]
        num_frames = len(aframes) #some might be empty
        ids = []
        d_arr = []
        for i in range(num_frames):
            num_bodies_in_frame = len(aframes[i][0][0])
            #print(fname, len(aframes[i][0][0]))
            if num_bodies_in_frame==1 and aframes[i][0]:
                #print(aframes[i][0][0][0])
                body_in_frame = aframes[i][0][0][0]
                ids.append(body_in_frame[0][0][0])
                #d_arr = aframes[i][0][0][1]
                #print(len(aframes[i][0][0][0][2]), len(aframes[i][0][0][0][2][0]), len(aframes[i][0][0][0][2][0][0]))
                joints_cords = body_in_frame[2][0]
                n_joints = len(joints_cords)
                j_cords = []
                for j in range(n_joints):
                    joint = joints_cords[j]
                    x,y,z = joint[0][0][0], joint[1][0][0], joint[2][0][0]
                    j_cords += [x,y,z]
                    #print(joint[0][0][0], joint[1][0][0], joint[2][0][0])
                d_arr.append(j_cords)
            elif num_bodies_in_frame>1:
                for k in range(num_bodies_in_frame):
                    #print(aframes[i][0][0][k][0])
                    body_in_frame = aframes[i][0][0][k]
                    ids.append(body_in_frame[0][0][0])
                    #d_arr = aframes[i][0][0][1]
                    #print(len(aframes[i][0][0][0][2]), len(aframes[i][0][0][0][2][0]), len(aframes[i][0][0][0][2][0][0]))
                    joints_cords = body_in_frame[2][0]
                    n_joints = len(joints_cords)
                    j_cords = []
                    for j in range(n_joints):
                        joint = joints_cords[j]
                        x,y,z = joint[0][0][0], joint[1][0][0], joint[2][0][0]
                        j_cords += [x,y,z]
                        #print(joint[0][0][0], joint[1][0][0], joint[2][0][0])
                    d_arr.append(j_cords)
        d_arr = np.array(d_arr)
        matpd = pd.DataFrame(d_arr)
        matpd['label'] = pd.DataFrame({'label':[label]*len(ids)})
        matpd['id'] = pd.DataFrame({'id':ids})
        #print('here', matpd)
        return matpd
        #
```

```python
import sys, os, shutil

lf = os.listdir('./data/A001/')
for f in lf:
    fname = './data/A001/'+f
    #print('processing '+fname)
    matfile2pandas(fname)
    #print('done')
```

```python
rdir='./data/'
matpd_all_actions = []
for i in range(60):
    dir_name = 'A0'
    if i<9:
        dir_name += '0'+str(i+1)
    else:
        dir_name += str(i+1)
    print('processing '+dir_name)
    d = rdir+dir_name + '/'
    lf = os.listdir(d)
    video_id = 0
    matpd_this_action=[]
    for f in lf:
        fname = d+f
        #print('processing '+fname)
        matpd = matfile2pandas(fname)
        if matpd is not None:
            #print(fname, matpd)
            matpd['video_id'] = pd.DataFrame({'video_id':[video_id]*len(matpd)})
            matpd_this_action.append(matpd)
            #print('done')
        video_id += 1
    matpd_this_action = pd.concat(matpd_this_action, ignore_index=True)
    matpd_all_actions.append(matpd_this_action)
    print('done ', dir_name)
```

```python
for i in range(60):
    action=i+1
    matpd_action = matpd_all_actions[i]
    fname = 'action_'+str(action)+'.csv'
    matpd_action.to_csv('./csv_data/'+fname, index=False)
```
